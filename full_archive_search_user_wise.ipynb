```python
def connect_to_endpoint(bearer_token, query,start_time, next_token=None):
    
    headers = {"Authorization":"Bearer {}".format(bearer_token)}
    
    # add additional parameters as needed
    params = {
            'query': query,
            'expansions': 'author_id,referenced_tweets.id,referenced_tweets.id.author_id,in_reply_to_user_id,entities.mentions.username',
            'tweet.fields': 'created_at,author_id,lang,entities,referenced_tweets,in_reply_to_user_id,public_metrics,context_annotations', 
            


            'start_time': start_time,
            'end_time': "2019-12-31T00:00:00Z",
            'max_results': 100,

        }
    url = "https://api.twitter.com/2/tweets/search/all"
    
    if (next_token is not None):       
        
        params['next_token'] = next_token # add pagination key to query dict   
        
    time.sleep(0.01)
    response = requests.request("GET", url, params=params, headers=headers)
    

    reset_time = response.headers.get('x-rate-limit-reset')
    if reset_time is not None:
        reset_time = int(reset_time)
        
    if response.status_code != 200:
        if(reset_time is not None):
            delay_time = reset_time - int(time.time())
        else:
            delay_time = 3
        print('Sleeping for '+str(delay_time)+' s')
        time.sleep(delay_time)
        response = requests.request("GET", url, params=params, headers=headers)
    if response.status_code != 200:
        raise Exception(response.status_code, response.text)
    
    return response.json()
```

```python
#Query based tweet crawling (Not user based)
############################################
purge = []
t = 0
fext = "all_tweets"
path_keywords = "drive/My Drive/MSR/ssr_keywords3/kerala_floods"
banned_users=[]
start_time = "2019-08-01T00:00:00Z"
c=0
#The keyword file does not exist. So, it has to be created
tweets = []
users = []
t += 1
start = time.time()
path = path_keywords+"/"

count = 0
flag = True
next_token = None

while flag:
    time.sleep(2)
    #If count of tweets collected exceeds this number (it can exceed quite a bit as result_count will already have count of all tweets in the current page)
    #if(count >=200000):  
    #    break
    #If there are no results or tweets in the next page, just collect tweets from this page
    if(next_token is None):
        json_response = connect_to_endpoint(bearer_token,query,start_time)
    #Else, send the next_token parameter in the connect function, so that next page can be referred to
    else:
        json_response = connect_to_endpoint(bearer_token,query,start_time, next_token)
    #Get the count of tweets in this page
    result_count = json_response['meta']['result_count']
    #Get the next token
    if 'next_token' in json_response['meta']:
        next_token = json_response['meta']['next_token']
    else:
        next_token = None
    #If there are more than 0 tweets in this page
    if result_count is not None and result_count > 0:
        #Append the tweets to a list
        #tweet_dict stands for one tweet
        #json_response['data'] contains all tweets in this page (number of tweets = result_count)
        for tweet_dict in json_response['data']:
            c+=1
            #print(tweet_dict.keys())
            tweets.append(tweet_dict)
            if c%1000 == 0:
              #Temporarily saving the partial tweets file (all tweets in the current page) for this user
              tweet_df2 = pd.DataFrame(tweets)
              tweet_df2.to_json(path+'kerala_floods.temp'+'.'+fext+'.'+str(start_time)+'.json')
              print(str(c)+" tweets collected.")
        #Append the users mentioned in the tweets (RT'ed etc.) to a list
        if('users' in json_response['includes'].keys()): #includes also returns an error key for invalid accounts/hashtags
            for tweet_dict in json_response['includes']['users']:
                users.append(tweet_dict)
        
        
        count += result_count
        #if(count%2000000==0):
        #    print("!!!!")
        #    print(count)
    #Else exit the while loop and go to the next user
    else:
        flag=False
    #If no more pages exist, exit the while loop and go to the next user
    if next_token is None:
        flag = False
    
    
tweet_df = pd.DataFrame(tweets)
#print(tweet_df)
tweet_df.to_json(path+'kerala_floods.'+fext+'.'+str(start_time)+'.json')
user_df = pd.DataFrame(users)
user_df.to_json(path+'kerala_floods_users'+'.'+fext+'.'+str(start_time)+'.json')
    
print((count,t))
```
